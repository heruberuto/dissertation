# ✂️ Update prompts from two files to single file

```python
import os
folder = os.path.join("prompts", "claimify")
prompts = {}
for filename in os.listdir(folder):
    if filename.endswith(".system.md") or filename.endswith(".user.md"):
        # pair system prompts with user prompts
        id, role = filename.split(".")[:2]
        if id not in prompts:
            prompts[id] = {}
        with open(os.path.join(folder, filename), "r") as f:
            prompts[id][role] = f.read()
```

```python
prompts
```

```python
for id, messages in prompts.items():
    # make file in folder id.md with their concat
    with open(os.path.join(folder, f"{id}.md"), "w") as f:
        f.write(messages['system'] + "\n\n# User message\n" + messages['user'])
```
