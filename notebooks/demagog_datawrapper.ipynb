```python
# load /mnt/data/factcheck/demagog/combined.ndjson as a dataframe
import pandas as pd
df = pd.read_json('/mnt/data/factcheck/demagog/combined.ndjson', lines=True)
```

```python
# make a subset that contains //datawrapper.dwcdn.net/ in assesment_explanation_html
df_subset = df[df['assessment_explanation_html_evidence_extracted'].str.contains('//datawrapper.dwcdn.net/', na=False)]
```

```python
# load this as df2
df2 = pd.read_json('/mnt/data/factcheck/demagog/demagog_statements.json')
```

```python
df2
```

```python
#filter where evidence one of the members contains //datawrapper.dwcdn.net/ in url field 
df2_subset = df2[df2['evidence'].apply(lambda x: any('datawrapper' in sub['url'] for sub in x) if isinstance(x, list) else False)]
```

```python
df2_subset
```

```python
# make a list of all datawrapper links in df_subset
datawrapper_links = []
for index, row in df2_subset.iterrows():
    for evidence in row['evidence']:
        if 'datawrapper' in evidence['url']:
            datawrapper_links.append(evidence['url'])
```

```python
# import url join function
from urllib.parse import urljoin

link_map = {l: urljoin(l, 'dataset.csv') for l in datawrapper_links}
link_map
```

```python
# try to fetch the csv files
import csv
import requests
import io

link_map_data = {}
failed = []
for link, csv_link in link_map.items():
    try:
        if csv_link.startswith('http://'):
            csv_link = csv_link.replace('http://', 'https://', 1)  # Ensure using https
        if csv_link.startswith('//'):
            csv_link = 'https:' + csv_link  # Add https: prefix if missing
        if csv_link.startswith('https://www.datawrapper.de/_/'):
            csv_link = csv_link.replace('https://www.datawrapper.de/_/', 'https://datawrapper.dwcdn.net/')
            csv_link=csv_link.replace('/dataset.csv','/1/dataset.csv')
        response = requests.get(csv_link)
        response.raise_for_status()  # Raise an error for bad responses
        df_csv = response.text
        #raise if empty text
        if not df_csv.strip():
            raise ValueError("Empty CSV content")
        link_map_data[link] = df_csv
    except Exception as e:
        print(f"Failed to fetch {csv_link}: {e}")
        try:
            csv_link = csv_link.replace("dataset.csv","data.csv")
            response = requests.get(csv_link)
            response.raise_for_status()  # Raise an error for bad responses
            df_csv = response.text
            if not df_csv.strip():
                raise ValueError("Empty CSV content")
            link_map_data[link] = df_csv
        except Exception as e:
            print(f"Failed to fetch {csv_link}: {e}")
            try:
                # data slug: 2ohiD for https://datawrapper.dwcdn.net/2ohiD/4/data.csv:
                # parse this slug
                parts = csv_link.split('/')
                slug = parts[3]
                csv_link = csv_link.replace("data.csv",f"data-{slug}.csv")
                response = requests.get(csv_link)
                response.raise_for_status()  # Raise an error for bad responses
                df_csv = response.text
                link_map_data[link] = df_csv
                if not df_csv.strip():
                    raise ValueError("Empty CSV content")
            except Exception as e:
                print(f"Failed to fetch {csv_link}: {e}")
                failed.append(csv_link)
```
