```python
# load /mnt/data/factcheck/demagog/combined.ndjson as a dataframe
import pandas as pd
```

```python
# load this as df2
df2 = pd.read_json('demagog_statements.json')
```

```python
df2
```

```python
#filter where evidence one of the members contains //datawrapper.dwcdn.net/ in url field 
df2_subset = df2[df2['evidence'].apply(lambda x: any('datawrapper' in sub['url'] for sub in x) if isinstance(x, list) else False)]
```

```python
df2_subset
```

```python
# make a list of all datawrapper links in df_subset
datawrapper_links = []
for index, row in df2_subset.iterrows():
    for evidence in row['evidence']:
        if 'datawrapper' in evidence['url']:
            datawrapper_links.append(evidence['url'])
```

```python
# import url join function
from urllib.parse import urljoin

link_map = {l: urljoin(l, 'dataset.csv') for l in datawrapper_links}
link_map
```

```python
import io


def parse_table(table_text):
    return table_text
    if "\t" in table_text:
        # tab separated
        df = pd.read_csv(io.StringIO(table_text), sep="\t")
    else:
        # comma separated
        df = pd.read_csv(io.StringIO(table_text))
    return df

import urllib.parse

def decode_octet(data_uri):
    # The raw data URI
    # 1. Isolate and decode the data part of the URI
    encoded_data = data_uri.split(',')[-1]
    decoded_data = urllib.parse.unquote(encoded_data)

    return decoded_data
```

```python
# try to fetch the csv files
import csv
import requests
from selenium import webdriver

driver = webdriver.Chrome()
link_map_data = {}
failed = []
for link, csv_link in link_map.items():
    try:
        if csv_link.startswith("http://"):
            csv_link = csv_link.replace("http://", "https://", 1)  # Ensure using https
        if csv_link.startswith("//"):
            csv_link = "https:" + csv_link  # Add https: prefix if missing
        if csv_link.startswith("https://www.datawrapper.de/_/"):
            csv_link = csv_link.replace("https://www.datawrapper.de/_/", "https://datawrapper.dwcdn.net/")
            csv_link = csv_link.replace("/dataset.csv", "/1/dataset.csv")
        response = requests.get(csv_link)
        response.encoding = response.apparent_encoding
        response.raise_for_status()  # Raise an error for bad responses
        df_csv = parse_table(response.text)
        # raise if empty text
        if not response.text.strip():
            raise ValueError("Empty CSV content")
        link_map_data[link] = df_csv
    except Exception as e:
        print(f"Failed to fetch {csv_link}: {e}")
        try:
            csv_link = csv_link.replace("dataset.csv", "data.csv")
            response = requests.get(csv_link)
            response.encoding = response.apparent_encoding
            response.raise_for_status()  # Raise an error for bad responses
            df_csv = parse_table(response.text)
            if not response.text.strip():
                raise ValueError("Empty CSV content")
            link_map_data[link] = df_csv
        except Exception as e:
            print(f"Failed to fetch {csv_link}: {e}")
            try:
                parts = csv_link.split("/")
                slug = parts[3]
                try:
                    csv_link = csv_link.replace("data.csv", f"data-{slug}.csv")
                    response = requests.get(csv_link)
                    response.encoding = response.apparent_encoding
                    response.raise_for_status()  # Raise an error for bad responses
                except:
                    csv_link = csv_link.replace(f"data-{slug}.csv", f"data-{slug}.csv".lower())
                    response = requests.get(csv_link)
                    response.encoding = response.apparent_encoding
                    response.raise_for_status()  # Raise an error for bad responses
                df_csv = parse_table(response.text)
                link_map_data[link] = df_csv
                if not response.text.strip():
                    raise ValueError("Empty CSV content")
            except Exception as e:
                print(f"Failed to fetch {csv_link}: {e}")
                try:
                    if True:
                        browser_url = "/".join(csv_link.split("/")[:-1])
                        driver.get(browser_url)
                        link_map_data[link] = decode_octet(driver.find_element("css selector", ".dw-data-link").get_attribute("href"))
                    else:
                        #remove the filename from csv_link
                        parts = csv_link.split("/")
                        csv_link = "/".join(parts[:-1])
                        response = requests.get(csv_link)
                        csv_part = ""
                        response.encoding = response.apparent_encoding
                        response.raise_for_status()  # Raise an error for bad responses
                        response_text = response.text
                        for line in response_text.splitlines():
                            stripped = line.strip()
                            lwr = stripped.lower()
                            if "chartdata" in lwr and ":" in lwr and not "ispreview" in lwr:
                                csv_part = stripped.split("chartData")[-1].strip().strip('",')
                                break
                            elif "data: \"" in lwr and "\\t" in lwr:
                                print("ahoj")
                                csv_part = stripped.split("data:")[-1].strip().strip('",')
                                break
                            
                        # raise if empty text
                        if not csv_part.strip():
                            raise ValueError("Empty CSV content")
                        link_map_data[link] = csv_part
                except Exception as e:
                    print(f"Failed to fetch {csv_link}: {e}")
                    failed.append(csv_link)
```
